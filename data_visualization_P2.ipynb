```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
f = r'C:\pandas\Toyota.csv'
cars_data = pd.read_csv(f,index_col=0,na_values=['??','????'])
```

```python
# removing missing values from the data frame.
cars_data.dropna(axis=0,inplace=True)
```

```python
# create a scatter plot of price vs age
sns.set(style="darkgrid")
sns.regplot(x=cars_data['Age'],y=cars_data['Price'])
```

```python
# we can remove the regression line by using.(fit_reg is by default true)
sns.regplot(x=cars_data['Age'],y=cars_data['Price'],fit_reg=False)
```

```python
# we can also change the apperiance by using keyword 'marker' .
sns.regplot(x=cars_data['Age'],y=cars_data['Price'],
            marker='*',fit_reg= False)
```

```python
# scatter plot of price vs age by fuel type .(we can add varibles also)
#using 'hue' to show fuel type category with diff color
sns.lmplot(x='Age',y='Price',data=cars_data,
           fit_reg=False,hue='FuelType',legend=True,palette='Set1')
```

```python
sns.displot(cars_data['Age'])
```

```python
sns.distplot(cars_data['Age'],kde=False)
```

```python
# fix the no of plots (intervals or range)
sns.histplot(cars_data['Age'],kde=False,bins=5)
```

```python
# bar plot ,frequency distribtion of fuel type of the class 
sns.countplot(x='FuelType',data=cars_data)
```

```python
# frequency distribtion of fuel type of the class with weathere the gear is A or M
sns.countplot(x="FuelType",data=cars_data,hue="Automatic")
```

```python
# box and wiskers plot 
sns.boxplot(y=cars_data['Price'])
```

```python
# check realtionship between two variables
# price of cars of various fuel types
sns.boxplot(x=cars_data['FuelType'],y=cars_data['Price'])
```

```python
sns.boxplot(x='FuelType',y='Price',hue='Automatic',data=cars_data)
```

```python
# box wiskers plot and histogram
# create 2 plots in same window
f, (ax_box, ax_hist) = plt.subplots(2, gridspec_kw={"height_ratios": [.15, .85]})
# now add create two plots
sns.boxplot(cars_data['Price'],ax=ax_box)
sns.distplot(cars_data['Price'],ax=ax_hist,kde=False)
```

```python
# pair wise plot / joint relationship 
sns.pairplot(cars_data,kind="Scatter",hue="FuelType")
plt.show()
```
