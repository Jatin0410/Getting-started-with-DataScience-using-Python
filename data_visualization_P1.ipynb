```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
file = 'C:\pandas\Toyota.csv'
cars_data = pd.read_csv(file,index_col=0,na_values=(["??","????"]))
```

```python
cars_data.dropna(axis=0,inplace=True)
```

```python
plt.scatter(cars_data['Age'], cars_data['Price'], c='red')
plt.title('Scatter plot of Price & Age of the cars.')
plt.xlabel('Age (months)')
plt.ylabel('Price (euros)')
plt.show()
```

```python
plt.hist(cars_data['KM'])
```

```python
plt.hist(cars_data['KM'],color='green',edgecolor='white',bins=5)
plt.title('Histogram of kilometers.')
plt.xlabel('Kilometer')
plt.ylabel('Frequency')
plt.show()
```

```python
counts=[979,120,12]
fuelType=('Petrol','Diesel','CNG')
index=np.arange(len(fuelType))
```

```python
plt.bar(index,counts,color=['red','blue','cyan'])
plt.title('Bar plot of fuel types.')
plt.xlabel('Fuel Type')
plt.ylabel('Frequency')
plt.xticks(index,fuelType,rotation=90)
```
